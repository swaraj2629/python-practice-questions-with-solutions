**Question 10:** Write a Python program to check if a variable is of a specific data type.

```python
# Your code here
l = ['Swaraj', 25, 5.9, 'Dudhe', True, None, 23+6j, [1, 2, 3], (4, 5), {'key': 'value'}]

for item in l:
    if type(item) == str:
        print(f"String: {item}")
    elif type(item) == int:
        print(f"Integer: {item}")
    elif type(item) == float:
        print(f"Float: {item}")
    elif type(item) == bool:
        print(f"Boolean: {item}")
    elif type(item) == type(None):
        print(f"NoneType: {item}")
    elif type(item) == complex:
        print(f"Complex Number: {item}")
    elif type(item) == list:
        print(f"List: {item}")
    elif type(item) == tuple:
        print(f"Tuple: {item}")
    elif type(item) == dict:
        print(f"Dictionary: {item}")        
    else:
        print(f"Unknown Type: {item}")
```

## 3. Basic Operators (Arithmetic, Comparison, Logical)

**Question 11:** Write a Python program to perform arithmetic operations: addition, subtraction, multiplication, and division.

```python
# Your code here
number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))
sum_result = number1 + number2
subtraction_result = number1 - number2
multiplication_result = number1 * number2
division_result = number1 / number2 if number2 != 0 else "Undefined (division by zero)"
print(f"Sum: {sum_result}, Subtraction: {subtraction_result}, Multiplication: {multiplication_result}, Division: {division_result}")
```

**Question 12:** Write a Python program to demonstrate comparison operators: equal to, not equal to, greater than, less than.

```python
# Your code here
number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))
equivalence = number1 == number2
not_equivalence = number1 != number2
greater_than = number1 > number2
less_than = number1 < number2
greater_equal = number1 >= number2
less_equal = number1 <= number2 
print(f"Equivalence: {equivalence}, Not Equivalence: {not_equivalence}, Greater Than: {greater_than}, Less Than: {less_than}, Greater Equal: {greater_equal}, Less Equal: {less_equal}")
```

**Question 13:** Write a Python program to demonstrate logical operators: and, or, not.

```python
# Your code here
number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))
logical_and = (number1 > 0) and (number2 > 0)
logical_or = (number1 > 0) or (number2 > 0)
logical_not = not (number1 > 0)
print(f"Logical AND: {logical_and}, Logical OR: {logical_or}, Logical NOT: {logical_not}")
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year = int(input("Enter the year: "))
if (year%4==0 and year%100!=0) or (year%400==0):
    print("Leap Year")
else:
    print("Not a leap year")
```

**Question 19:** Write a Python program to check if a string is a palindrome.

**Question 20:** Write a Python program to sort a list of numbers in ascending order.

```python
# Your code here
n = int(input("Enter a how numbers you want in list? : "))
l1 = []
for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    l1.append(num)
print(f"The list of numbers is: {l1}")
print("the sorted list is:", sorted(l1))
```
